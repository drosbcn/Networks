```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from graphviz import Digraph

file ='ICIO_data/ICIO2016_2011.csv'
norm_cutoff = 0.01
# Import the input-output table, with multi-indices for both index and columns: one for country, another for sector
input_output = pd.read_csv(file, index_col = [0,1], header = [0,1])
```

```python
input_output
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        total = df[feature_name].sum()
        if total == 0:
            result = result.drop(feature_name,1)
            result = result.drop(feature_name,0)
        else: 
            result[feature_name] = df[feature_name] / total
    return result 
```

```python
input_output_norm = normalize(input_output)
input_output_norm[input_output_norm<norm_cutoff] = 0

input_output_matrix = input_output_norm.as_matrix()
G = nx.from_numpy_matrix(input_output_matrix)

#make a filter for several countries
smpl_cntrs = ['FRA','GBR']
index_sample = input_output_norm.loc[smpl_cntrs]
col_idx_sample = index_sample[smpl_cntrs]
```

```python
#filter input_output by index, extract 
#degree centrality

deg_cent = nx.degree_centrality(G)

plt.hist(deg_cent.values(), bins = "auto")
plt.title('Degree Centrality')

plt.show()
```

```python
cent_rank = []
for i , val in enumerate(list(input_output_norm.index)):
    a = list(val)
    b = deg_cent.values()[i]
    a.append(b)
    cent_rank.append(a)
cent_rank = pd.DataFrame(cent_rank)
cent_rank.columns = ['Country', 'Sector', 'Degree Centrality']
cent_rank.sort_values(by = 'Degree Centrality', ascending = False).head(10)
```

```python
#eigenvector centrality
eigen_vec_centrality = nx.eigenvector_centrality(G, max_iter=100, weight='weight')
plt.hist(eigen_vec_centrality.values(), bins="auto")
plt.title('Eigenvector Centraliry')
plt.show()
```

```python
eigen_cent_rank = []
for i , val in enumerate(list(input_output_norm.index)):
    a = list(val)
    b = eigen_vec_centrality.values()[i]
    a.append(b)
    eigen_cent_rank.append(a)
eigen_cent_rank = pd.DataFrame(eigen_cent_rank)
eigen_cent_rank.columns = ['Country', 'Sector', 'Eigenvector Centrality']
eigen_cent_rank.sort_values(by = 'Eigenvector Centrality', ascending = False).head(10)
```

```python
#betweeness centrality
btw_centrality = nx.betweenness_centrality(G)
plt.hist(btw_centrality.values(), bins="auto")
plt.title('betweeness centrality')
plt.show()
```

```python
btw_cent_rank = []
for i , val in enumerate(list(input_output_norm.index)):
    a = list(val)
    b = btw_centrality.values()[i]
    a.append(b)
    btw_cent_rank.append(a)
btw_cent_rank = pd.DataFrame(btw_cent_rank)
btw_cent_rank.columns = ['Country', 'Sector', 'Betweeness Centrality']
btw_cent_rank.sort_values(by = 'Betweeness Centrality', ascending = False).head(10)
```
