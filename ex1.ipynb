```python
import matplotlib.pyplot as plt
import networkx as nx

def er_graph(n,p,d=False):
    G = nx.erdos_renyi_graph(n,p,directed=d)
    nx.draw(G)
    plt.title("Erdos - Renyi random graph, prob = " + str(p))
    plt.plot()
    # plt.savefig()
    # plt.show()


def er_degree_distr(n,p):
    G = nx.erdos_renyi_graph(n,p)
    degree_sequence=list(nx.degree(G).values())
    plt.hist(degree_sequence,bins =100)
    plt.title("Degree distribution plot, prob = " + str(p))
    plt.plot()
    #plt.show()

P = [0.01, 0.1, 0.25, 0.6, 0.9]
```

```python
plt.figure(figsize=(10,13))
for i,p in enumerate(P,1):
    plt.subplot(3,2,i)
    er_graph(50,p)
plt.show()
```

```python
plt.figure(figsize=(13,13))
plt.ylabel("frequency")
plt.xlabel("degree")
for i,p in enumerate(P,1):
    plt.subplot(3,2,i)
    er_degree_distr(5000,p)
plt.show()
```
