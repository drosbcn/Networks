```python
import matplotlib.pyplot as plt
import networkx as nx

#pos = pos (dictionary) #â€“ A dictionary with nodes as 
#keys and positions as values. Positions should be sequences of length 2.

def er_graph(n,p,d=False):
    G = nx.erdos_renyi_graph(n,p,directed=d)
#    nx.draw(G)
    pos=nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos,node_color='b', alpha = 0.3)
    nx.draw_networkx_edges(G, pos)
    plt.title("Erdos - Renyi random graph, prob = " + str(p))
    plt.plot(alpha=.005)
    # plt.savefig()
    # plt.show()


def er_degree_distr(n,p):
    G = nx.erdos_renyi_graph(n,p)
    degree_sequence=list(nx.degree(G).values())
    plt.hist(degree_sequence,bins =100)
    plt.title("Degree distribution plot, prob = " + str(p))
    plt.plot()
    #plt.show()

P = [0.01, 0.1, 0.25, 0.6, 0.9]
```

```python
plt.figure(figsize=(10,13))
for i,p in enumerate(P,1):
    plt.subplot(3,2,i)
    er_graph(50,p)
plt.show()
```

Commentary:
As we can see from the plots, the lower probability given prodives us with less connections. Based on the statement that a fixed number of edges is likely, as the probability of creating another edge with a different node increases. With high probability we get most of the nodes interconnected with each other leaving 2 vertexes on  
either side.

```python
plt.figure(figsize=(13,13))
plt.ylabel("frequency")
plt.xlabel("degree")
for i,p in enumerate(P,1):
    plt.subplot(3,2,i)
    er_degree_distr(5000,p)
plt.show()
```

```python
#calculating average path size and average number of components for all probabilities
path_size = []
com_len = []
for p in P:
    G = nx.erdos_renyi_graph(5000,p)
    avg_path = nx.average_shortest_path_length(G)
    con_com = nx.number_connected_components(G)
    path_size.append(avg_path)
    com_len.append(con_com)
    print("average path size for p=" + str(p) + str(avg_path))
    print("average number of connected components for p="+ str(p) + str(con_com))
```
